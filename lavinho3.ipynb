```python
class shape:
    def area(a):
        return 0
class rectangle(shape):
    def __init__(a, width, length):
        a.length = length
        a.width = width
    def area(a):
        return a.length * a.width
if __name__ == "__main__":
    rectangle = rectangle(5, 7)
    print(rectangle.area())
```

```python
import math
class Point:
    def __init__(a, x, y):
        a.x = x
        a.y = y
    def show(a):
        print(a.x, a.y)
    def move(a, new_x, new_y):
        a.x = new_x
        a.y = new_y
    def dist(a, other_point):
        return math.sqrt((a.x - other_point.x) ** 2 + (a.y - other_point.y) ** 2)
if __name__ == "__main__":
    p1 = Point(3, 4)
    p2 = Point(6, 8)
    p1.show()
    p2.show()
    print(p1.dist(p2))
```

```python
class BankAccount:
    def __init__(a, owner, balance=0):
        a.owner = owner
        a.balance = balance
    def deposit(a, amount):
        if amount > 0:
            a.balance += amount
            print("Deposit successful. New balance:", a.balance)
        else:
            print("Deposit amount must be positive.")
    def withdraw(a, amount):
        if 0 < amount <= a.balance:
            a.balance -= amount
            print("Withdrawal successful. New balance:", a.balance)
        else:
            print("Insufficient funds or invalid amount.")
if __name__ == "__main__":
    acc = BankAccount("Akezhan", 1000)
    acc.deposit(500)
    acc.withdraw(200)
    acc.withdraw(1500)
```

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
nums = [10, 15, 3, 7, 19, 23, 4, 8, 29]
prime_nums = list(filter(lambda x: is_prime(x), nums))
prime_nums.sort()
print(prime_nums)
```

```python
def filter_prime(a):
    def is_prime(x):
        if x < 2:
            return False
        for i in range(2, int(x ** 0.5) + 1):
            if x % i == 0:
                return False
        return True
    return [x for x in a if is_prime(x)]
a = [1, 10, 3, 7, 9, 32]
print(filter_prime(a))
```

```python
def print_permutations(a, prefix=""):
    if len(a) == 0:
        print(prefix)
    else:
        for i in range(len(a)):
            print_permutations(a[:i] + a[i+1:], prefix + a[i])
s = input()
print_permutations(s)
```

```python
def has_33(nums):
    for i in range(len(nums) - 1):
        if nums[i] == 3 and nums[i+1] == 3:
            return True
    return False        
my_list = list(map(int,input().split()))
print(has_33(my_list))
```

```python
def has_007(nums):
    need = [0,0,7]
    for i in nums:
        if i == need[0]: 
            need.pop(0)
        if len(need) == 0:   
            return True
    return False        
my_list = list(map(int,input().split()))
print(has_007(my_list))

```

```python
def uniques_only(listo):
    unique_list = []
    for i in listo:
        if i not in unique_list:
            unique_list.append(i)
    return unique_list    
li = list(map(int,input().split()))
print(uniques_only(li))
```

```python
def is_polindrome(word):
    reversed_w = ""
    for i in word[::-1]:
        reversed_w += i
    if word == reversed_w:
        return True
    else:
        return False
word = str(input())
print(is_polindrome(word))
```

```python
def histo(listo):
    for i in listo:
        iterats = i
        res = ""
        while iterats != 0:
            res += '*'
            iterats -= 1
        print(res)    
li = list(map(int,input().split()))
histo(li)
```

```python
import random
def guess_number():
    name = str(input("Hello! What is your name?"))
    print(f"Well, {name}, I am thinking of a number between 1 and 20.")    
    number_g = random.randint(1,20)
    attemps = 0
    while True:
        try:
            guess_n = int(input("Take a guess."))
            attemps += 1
            if guess_n < number_g:
                print("Your guess is too low.")
            elif guess_n > number_g: 
                print("Your guess is too high.")
            else:
                print(f"Good job, {name}! You guessed my number in {attemps} guesses!")
                break
        except ValueError:
            print("Please enter a valid number between 1 and 20.")

guess_number()
```

```python
def score(movie):
    return movie.get('imdb', 0) > 5.5
    
def score_movies(movies):
    score_list = []
    for movie in movies:
        if score(movie):
            score_list.append(movie)
    return score_list
print(score_movies(movies))
```

```python
def average_score(movies):
    total_score = 0
    for movie in movies:
        total_score += movie.get('imdb', 0)
    return total_score / len(movies) if movies else 0
print("avg is " + str(average_score(movies)))  
```
